#準備

```python

import numpy as np
import pandas as pd
# from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import japanize_matplotlib
import seaborn as sns
```

# データをファイルから読み込む

エクセルファイルをPython環境に読み込むためにPandasの
read_excel()というメソッドを使う


```python
# 読み込むファイルを指定する
file_path = 'customer_data.xlsx'

# df という変数に読み込む
df = pd.read_excel(file_path)
```

```python
df.info()
```

#ヒストグラム

```python
import matplotlib.pyplot as plt
import japanize_matplotlib
import seaborn as sns

# 年齢のデータを取得
ages = df['年齢']

# ヒストグラムを描画
sns.histplot(ages, bins=20, kde=False)
plt.title('年齢の分布')
plt.xlabel('年齢')
plt.ylabel('人数')
plt.show()
```

# 件数

```python
# 職業のデータを取得
sns.countplot(x='職業', data=df)
plt.show()
```

```python
# 職業のデータを取得
sns.countplot(y='職業', data=df)
plt.show()
```

```python
# 職業のデータを取得
sns.countplot(x='職業', hue='性別', data=df)
plt.show()
```

```python
sns.countplot(x='性別', hue='職業', data=df, order=['女', '男'], hue_order=['学生', '会社員', '公務員', 'その他'])
plt.show()
```

```python
#sns.catplotで箱ひげ図を描く
sns.catplot(x='性別', y='年収', data=df, kind='box')
plt.show()
```

```python
sns.jointplot(x='Aの利用回数', y='Bの利用回数', data=df, hue='年齢')
plt.show()
```

# 散布図の表示

```python
# Plot scatter plot for Age vs Income using seaborn
sns.scatterplot(data=df, x='年齢', y='年収', hue='性別', palette={'男': 'blue', '女': 'red'})
plt.title('Scatter Plot of Age vs Income')
plt.show()
```
